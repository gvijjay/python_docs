```python
# Load employees and salaries again
import pandas as pd
employees = pd.read_csv('employees.csv')
salaries = pd.read_csv('salaries.csv')
departments = pd.read_csv('departments.csv')
```

```python
# 1. Line plot of salaries
import matplotlib.pyplot as plt
plt.plot(employees['EmpID'], employees['Salary'])
plt.title("Salary vs Employee ID")
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.show()
```

```python
# 2. Bar chart of average salary per department
employees.groupby('Dept')['Salary'].mean().plot(kind='bar', title="Avg Salary per Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
# 3. Histogram of salaries
plt.hist(employees['Salary'], bins=5, color='skyblue', edgecolor='black')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Horizontal bar chart of employee names vs salaries
plt.barh(employees['Name'], employees['Salary'], color='green')
plt.title("Employee Salaries")
plt.xlabel("Salary")
plt.ylabel("Name")
plt.show()
```

```python
# 5. Pie chart of department distribution
employees['Dept'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Department Distribution")
plt.ylabel("")
plt.show()
```

```python
# 6. Line plot of bonus with markers(plots customization)
plt.plot(salaries['EmpID'], salaries['Bonus'], marker='o', linestyle='--', color='purple')
plt.title("Bonus by Employee ID")
plt.xlabel("EmpID")
plt.ylabel("Bonus")
plt.grid(True)
plt.show()
```

```python
# 7. Bar chart with custom colors and legend
avg_salary = employees.groupby('Dept')['Salary'].mean()
avg_salary.plot(kind='bar', color=['red', 'blue', 'orange'])
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.legend(["Avg Salary"])
plt.show()
```

```python
# 8. Stacked bar chart of Salary and Bonus
merged = pd.merge(employees, salaries, on='EmpID')
plt.bar(merged['Name'], merged['Salary'], label='Salary')
plt.bar(merged['Name'], merged['Bonus'].fillna(0), bottom=merged['Salary'], label='Bonus')
plt.title("Total Compensation")
plt.xlabel("Employee")
plt.ylabel("Amount")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Customized histogram with grid
plt.hist(salaries['Tax'].dropna(), bins=4, color='lightcoral', edgecolor='black')
plt.title("Tax Distribution")
plt.xlabel("Tax")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# 10. Scatter plot of Salary vs Bonus
plt.scatter(merged['Salary'], merged['Bonus'], color='teal')
plt.title("Salary vs Bonus")
plt.xlabel("Salary")
plt.ylabel("Bonus")
plt.show()
```

```python
#Subplots
# 11. Two line plots in one figure
plt.plot(employees['EmpID'], employees['Salary'], label='Salary')
plt.plot(salaries['EmpID'], salaries['Bonus'], label='Bonus')
plt.title("Salary and Bonus Comparison")
plt.xlabel("EmpID")
plt.legend()
plt.show()
```

```python
# 12. Subplot: salary and bonus histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(employees['Salary'], color='skyblue')
axs[0].set_title("Salary Distribution")
axs[1].hist(salaries['Bonus'].dropna(), color='salmon')
axs[1].set_title("Bonus Distribution")
plt.tight_layout()
plt.show()
```

```python
# 13. Subplot: salary and tax line plots
fig, axs = plt.subplots(2, 1)
axs[0].plot(employees['EmpID'], employees['Salary'], color='blue')
axs[0].set_title("Salary by EmpID")
axs[1].plot(salaries['EmpID'], salaries['Tax'].fillna(0), color='red')
axs[1].set_title("Tax by EmpID")
plt.tight_layout()
plt.show()
```

```python
# 14. Bar and line plot in same figure
fig, ax1 = plt.subplots()
ax1.bar(employees['Name'], employees['Salary'], color='grey', label='Salary')
ax2 = ax1.twinx()
ax2.plot(employees['Name'], employees['Salary'].cumsum(), color='blue', label='Cumulative')
ax1.set_title("Salary and Cumulative Salary")
ax1.set_xlabel("Employee")
ax1.set_ylabel("Salary")
ax2.set_ylabel("Cumulative")
fig.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 15. Boxplot of salaries
plt.boxplot(employees['Salary'])
plt.title("Salary Spread")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
# 16. Area chart of cumulative salary
employees_sorted = employees.sort_values('EmpID')
plt.fill_between(employees_sorted['EmpID'], employees_sorted['Salary'].cumsum(), color='lightgreen', alpha=0.6)
plt.title("Cumulative Salary (Area Chart)")
plt.xlabel("EmpID")
plt.ylabel("Cumulative Salary")
plt.show()
```

```python
# 17. Donut chart of department distribution
dept_counts = employees['Dept'].value_counts()
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4})
plt.title("Department Share (Donut Chart)")
plt.show()
```

```python
# 18. Step plot for salary trend
plt.step(employees['EmpID'], employees['Salary'], where='mid', color='brown')
plt.title("Step Plot of Salaries")
plt.xlabel("EmpID")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
# 19. Error bar plot (assuming std dev = 2000 for demo)
plt.errorbar(employees['EmpID'], employees['Salary'], yerr=2000, fmt='o', capsize=5, label='Salary with Error')
plt.title("Salary with Error Bars")
plt.xlabel("EmpID")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# 20. Heatmap of correlation between salary, bonus, and tax
import seaborn as sns
merged = pd.merge(employees, salaries, on='EmpID')
sns.heatmap(merged[['Salary', 'Bonus', 'Tax']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 21. Hexbin plot for Salary vs Bonus (useful for dense data)
plt.hexbin(merged['Salary'], merged['Bonus'].fillna(0), gridsize=10, cmap='Blues')
plt.colorbar(label='Count')
plt.title("Hexbin: Salary vs Bonus")
plt.xlabel("Salary")
plt.ylabel("Bonus")
plt.show()
```

```python
# 22. Violin plot of Salary (needs seaborn)
sns.violinplot(y=employees['Salary'], color='lightblue')
plt.title("Violin Plot of Salary Distribution")
plt.show()
```

```python
# 25. Count plot of departments
sns.countplot(x='Dept', data=employees)
plt.title("Employee Count by Department")
plt.show()
```
